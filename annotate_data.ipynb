```python
import spacy
from spacy.training import Example
import random

nlp = spacy.blank("en")  # Start with a blank English model
ner = nlp.add_pipe("ner")

# Add the SKILL label
ner.add_label("SKILL")

# Prepare training data
examples = [Example.from_dict(nlp.make_doc(text), annotations) for text, annotations in TRAIN_DATA]


```

```python
# Train
optimizer = nlp.begin_training()
for i in range(30):
    random.shuffle(examples)
    print("ma")
    losses = {}
    for example in examples:
        nlp.update([example], drop=0.3, sgd=optimizer, losses=losses)
    print(f"Iteration {i+1}, Losses: {losses}")

# Save model
nlp.to_disk("skill_ner_model")
```
