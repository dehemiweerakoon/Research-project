```python
import cloudscraper
from bs4 import BeautifulSoup

medium_url = "https://medium.com/@Dehemiweerakoon"

scraper = cloudscraper.create_scraper()
html = scraper.get(medium_url).text

soup = BeautifulSoup(html, "html.parser")
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

# 1. Define your long text
text = """
This is Paragraph 1. It has some detailed information about a topic.
It is long enough that we might want to keep it together.

This is Paragraph 2. It is separated by double newlines. 
If our chunk size is small, the splitter will break this into pieces.
"""

# 2. Initialize the splitter
# chunk_size: Max characters per chunk
# chunk_overlap: How many characters to repeat from the previous chunk
splitter = RecursiveCharacterTextSplitter(
    chunk_size=100,
    chunk_overlap=20,
    separators=["\n\n", "\n", " ", ""]
)

# 3. Create the chunks
chunks = splitter.split_text(text)

# 4. Display results
print(f"Total chunks created: {len(chunks)}\n")
for i, chunk in enumerate(chunks):
    print(f"--- Chunk {i+1} ({len(chunk)} chars) ---")
    print(chunk)

```
